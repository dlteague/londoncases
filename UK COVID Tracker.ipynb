```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
deaths = pd.read_csv("https://assets.publishing.service.gov.uk/government/uploads/system/uploads/attachment_data/file/887263/COVID-19_-_time_series_of_daily_deaths_25_May_2020.csv", parse_dates=['Publicly confirmed as deceased as of 5pm this day'], index_col=['Publicly confirmed as deceased as of 5pm this day'])
cases = pd.read_csv("https://c19downloads.azureedge.net/downloads/csv/coronavirus-cases_latest.csv", parse_dates=['Specimen date'], index_col=['Specimen date'])
```

```python
ax = deaths.plot (y=1,logy=True,legend=False, figsize=(15,5))
#ax.set_xticklabels([t if not i%2 else "" for i,t in enumerate(ax.get_xticklabels())])
```

```python
bx = deaths.plot.bar(x=0, y=2,rot=30,legend=False,figsize = (15,5))
bx.set_xticklabels([t if not i%10 else "" for i,t in enumerate(bx.get_xticklabels())]);
```

```python
dailycases = cases.resample('D').sum()
cx = dailycases.plot.bar(y=0,rot=30,legend=False,figsize=(15,5))
cx.set_xticklabels([pandas_datetime.strftime("%d-%b-%y") for pandas_datetime in dailycases.index])
cx.set_xticklabels([t if not i%10 else "" for i,t in enumerate(cx.get_xticklabels())]);
```

```python
dailycases.plot(y=2,logy=True,legend=False,figsize=(15,5))
```

```python
londoncases = cases[cases['Area name']=="London"]
dailylondoncases = londoncases.resample('D').sum()
cx = dailylondoncases.plot.bar(y=0,rot=30,legend=False,figsize=(15,5))
cx.set_xticklabels([pandas_datetime.strftime("%d-%b-%y") for pandas_datetime in dailylondoncases.index])
cx.set_xticklabels([t if not i%10 else "" for i,t in enumerate(cx.get_xticklabels())]);
```
